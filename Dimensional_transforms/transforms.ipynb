```python
import matplotlib.pyplot as plt
import numpy as np

def plot_points(points, title):
    x, y = zip(*points)
    plt.scatter(x, y, label=title)

def translate(points, dx, dy):
    return [(x + dx, y + dy) for x, y in points]

def scale(points, sx, sy):
    return [(x * sx, y * sy) for x, y in points]

def reflect_x(points):
    return [(x, -y) for x, y in points]

def shear_x(points, shx):
    return [(x + shx * y, y) for x, y in points]

# Define a set of points
original_points = [(0, 0), (1, 1), (2, 2), (3, 3)]

# Create a figure and axis for plotting
fig, ax = plt.subplots()

# Plot the original points
plot_points(original_points, "Original")

# Perform transformations
translated = translate(original_points, 2, 3)
scaled = scale(original_points, 2, 0.5)
reflected = reflect_x(original_points)
sheared = shear_x(original_points, 1)

# Plot the transformed points
plot_points(translated, "Translated")
plot_points(scaled, "Scaled")
plot_points(reflected, "Reflected")
plot_points(sheared, "Sheared")

# Set the aspect ratio for the plot
ax.set_aspect('equal', adjustable='box')

# Add legend
plt.legend()

# Show the plot
plt.grid()
plt.title("2D Transformations in Computer Graphics")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_points(points, title):
    x, y = zip(*points)
    plt.scatter(x, y, label=title)

def rotate(points, angle_deg):
    angle_rad = np.radians(angle_deg)
    rotation_matrix = np.array([[np.cos(angle_rad), -np.sin(angle_rad)],
                                [np.sin(angle_rad), np.cos(angle_rad)])
                                
    return [(np.dot(rotation_matrix, np.array([x, y])).tolist()) for x, y in points]

if __name__ == "__main__":
    # Define a set of points for an object to be rotated
    original_points = [(1, 1), (2, 2), (3, 3), (4, 4)]

    # Create a figure and axis for plotting
    fig, ax = plt.subplots()

    # Plot the original object
    plot_points(original_points, "Original")

    # Define the rotation angle in degrees
    rotation_angle = 45

    # Perform the rotation
    rotated_points = rotate(original_points, rotation_angle)

    # Plot the rotated object
    plot_points(rotated_points, "Rotated")

    # Set the aspect ratio for the plot
    ax.set_aspect('equal', adjustable='box')

    # Add legend
    plt.legend()

    # Show the plot
    plt.grid()
    plt.title("2D Rotation in Computer Graphics")
    plt.show()
```
