```python
import matplotlib.pyplot as plt

def bresenham_circle(radius):
    x = 0
    y = radius
    p = 3 - 2 * radius
    
    points = []
    
    while x <= y:
        points.extend([
            (x, y), (y, x), (-x, y), (-y, x),
            (-x, -y), (-y, -x), (x, -y), (y, -x)
        ])
        
        x += 1
        
        if p < 0:
            p = p + 4 * x + 6
        else:
            y -= 1
            p = p + 4 * (x - y) + 10
    
    return points
    
```

```python
def plot_points(points):
    x_coords, y_coords = zip(*points)
    plt.scatter(x_coords, y_coords, color='red', s=10)
    
```

```python
def main():
    radius = 10
    circle_points = bresenham_circle(radius)
    
    plt.title("Bresenham's Circle Drawing")
    plot_points(circle_points)
    
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True)
    plt.show()
    
```

```python
if __name__ == "__main__":
    main()
    
```
