```python
import matplotlib.pyplot as plt

def draw_ellipse(a, b):
    # Initialize parameters
    x, y = 0, b  # Starting point

    # Region 1
    p1 = (b * b) - (a * a * b) + (0.25 * a * a)
    dx = 2 * b * b * x
    dy = 2 * a * a * y

    # Plot the initial point in all regions
    plot_points(x, y)

    # Region 1
    while dx < dy:
        x += 1
        if p1 < 0:
            dx += 2 * b * b
            p1 += dx + b * b
        else:
            y -= 1
            dx += 2 * b * b
            dy -= 2 * a * a
            p1 += dx - dy + b * b
        plot_points(x, y)

    # Region 2
    p2 = (b * b * (x + 0.5) * (x + 0.5)) + (a * a * (y - 1) * (y - 1)) - (a * a * b * b)
    while y > 0:
        y -= 1
        if p2 > 0:
            dy -= 2 * a * a
            p2 += a * a - dy
        else:
            x += 1
            dx += 2 * b * b
            dy -= 2 * a * a
            p2 += dx - dy + a * a
        plot_points(x, y)

def plot_points(x, y):
    # Plot points (x, y) and their symmetric counterparts in all 4 quadrants
    plt.scatter(x, y)
    plt.scatter(-x, y)
    plt.scatter(x, -y)
    plt.scatter(-x, -y)

if __name__ == "__main__":
    a = 10  # Semi-major axis
    b = 5   # Semi-minor axis

    # Create a figure and axis for plotting
    fig, ax = plt.subplots()

    # Call the function to draw the ellipse
    draw_ellipse(a, b)

    # Set equal aspect ratio to display a proper ellipse
    ax.set_aspect('equal', adjustable='box')

    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Midpoint Ellipse Algorithm")

    # Show the plot
    plt.grid()
    plt.show()
```
