```python
import matplotlib.pyplot as plt
```

```python
def plot_circle_points(center_x, center_y, x, y):
    plt.plot(center_x + x, center_y + y, 'ro')
    plt.plot(center_x + y, center_y + x, 'ro')
    plt.plot(center_x - x, center_y + y, 'ro')
    plt.plot(center_x - y, center_y + x, 'ro')
    plt.plot(center_x + x, center_y - y, 'ro')
    plt.plot(center_x + y, center_y - x, 'ro')
    plt.plot(center_x - x, center_y - y, 'ro')
    plt.plot(center_x - y, center_y - x, 'ro')
```

```python
def midpoint_circle(center_x, center_y, radius):
    x = 0
    y = radius
    p = 1 - radius
    
    plot_circle_points(center_x, center_y, x, y)
    
    while x < y:
        x += 1
        if p < 0:
            p += 2 * x + 1
        else:
            y -= 1
            p += 2 * (x - y) + 1
        
        plot_circle_points(center_x, center_y, x, y)
```

```python
def main():
    center_x = 0
    center_y = 0
    radius = 5
    
    plt.figure(figsize=(6, 6))
    plt.title("Midpoint Circle Generation Algorithm")
    
    midpoint_circle(center_x, center_y, radius)
    
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True)
    plt.show()
```

```python
if __name__ == "__main__":
    main()
```
